```python
#Zieba Glaz Szymon Artur

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ejercicio 2

t = sp.symbols('t')

expr1 = 50*sp.E**(0.3*t)
expr2 = 6*sp.E**(0.5*t)
f1 = sp.Lambda(t, expr1)
f2 = sp.Lambda(t, expr2)
sp.pprint(f1(t))
sp.pprint(f2(t))

soluciones1 = sp.solve(f1(t) - 1000, t)
soluciones2 = sp.solve(f1(t) - f2(t), t)

print("\nEjericio 2.a\n")

for i in range(len(soluciones1)):
    print(f"solucion {i + 1} = {soluciones1[i]}")

print("\nEjericio 2.b\n")

for i in range(len(soluciones2)):
    print(f"solucion {i + 1} = {soluciones2[i]}")
```

```python
# Ejercicio 3

t = sp.symbols('t')

g = 9.8
a = 1

expr = a*(1 - sp.E**(-g*t/a))
f = sp.Lambda(t, expr)

sp.pprint(f(t))

soluciones = sp.solve(f(t) - 0.99*a, t)

print("\nEjericio 3\n")

print(f"solucion {1} = {soluciones[0]}")

# La funcion no tiene asindotas vertiales ya que es un exponencial restado de 1 y multiplicado por a que es una constante la cual es continua en todo R

print(f"resultado por la derecha: {sp.limit(f(t), t, sp.oo)}")
print(f"resultado por la izquierda: {sp.limit(f(t), t, -sp.oo)}")

m = sp.limit(f(t)/t, t, sp.oo)
n = sp.limit(f(t) - m*t, t, sp.oo)

print(f"resultado de m: {m}")
print(f"resultado de n: {n}")

print(sp.limit(f(t) - (m*t +n), t, sp.oo))
print(sp.limit(f(t) - (m*t +n), t, -sp.oo))

# Plot the function f(t)
t_vals = np.linspace(-1, 100, 5000)
f_vals = [f(t_val) for t_val in t_vals]

plt.plot(t_vals, f_vals, label="f(t)")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.title("Plot of f(t)")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.legend()
plt.grid()
plt.show()
```
