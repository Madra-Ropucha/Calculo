```python
import sympy as sp
from sympy import latex
import numpy as np
from IPython.display import display, Math

from tema2 import NewtonRaphson
```

```python
def Derivada(f, var):
    h = sp.symbols('h')
    return sp.limit((f.subs(var, var + h) - f) / h, h, 0)
```

```python
x = sp.symbols('x')
f = x**3 + 2*x**2 + x + 1
df = Derivada(f, x)
print(f"Función: {f}")
print(f"Derivada: {df}")
```

```python
x = sp.symbols('x')
f1Expr = (1 - sp.E**(-x)) / x
f2Expr = 1

f1Diff = sp.diff(f1Expr, x)
f2Diff = sp.diff(f2Expr, x)

print(sp.limit(f1Diff, x, 0, dir='-'))
print(sp.limit(f2Diff, x, 0))
print(sp.limit(f1Diff, x, 0, dir='+'))
```

```python
x = sp.symbols('x')
fExpr = x**3 + 2*x - 2

NewtonRaphson(fExpr, 0.5,  maxIter=10, tol=1e-6)
```

```python
x = sp.symbols('x')
fExpr = sp.atan(sp.sin(x) / (1 + sp.cos(x)))


display(Math('f(x) = ' + latex(fExpr)))
```

```python
x = sp.symbols('x')
fExpr = (4*x + 1)**(2 + sp.sin(x**2))

fExpr

sp.diff(fExpr, x)
```

```python
x = sp.symbols('x')
fExpr = sp.Piecewise(
    ((sp.sin(3*x)**2) / (3*x**2), x != 0),
    (sp.Rational(1, 3), x == 0)
)

fExpr
```

```python
x = sp.symbols('x')
fExpr = sp.Piecewise(
    (sp.sin(x), x < 0),
    (sp.atan(x), x <= 1),
    (sp.pi*x**3/4, True)
)

fExpr
```

```python
x = sp.symbols('x')
fExpr = x**3*sp.E**x

fExpr

fDiff = sp.diff(fExpr, x)

print(sp.solve(fDiff, x))
print(fDiff.subs(x, -4))
print(fDiff.subs(x, -1))
print(fDiff.subs(x, 1))
```
