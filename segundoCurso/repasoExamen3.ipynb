```python
# MÃ©todos simples de integraciÃ³n numÃ©rica (hecha por udc)

import sympy as sp
import numpy as np

def puntoMedio(a, b, fpm):
    aprox_pm = (b-a) * fpm
    return aprox_pm

def trapecio(a, b, fa, fb):
    aprox_tr = (b-a) * (fa + fb)/2
    return aprox_tr

def simpson(a, b, fa, fpm, fb):
    aprox_simp = (b-a) * (fa + 4*fpm + fb)/6
    return aprox_simp

x = sp.Symbol('x', real = True)

f_exp = x**4 + 1
f = sp.lambdify(x,f_exp)

a = 0
b = 3
pm = (a+b)/2

fa = f(a)
fpm = f(pm)
fb = f(b)

print('Valor aproximado de I mediante la fÃ³rmula del punto medio = ', puntoMedio(a,b,fpm) ) 
print('Valor aproximado de I mediante la fÃ³rmula del trapecio = ', trapecio(a,b,fa,fb) ) 
print('Valor aproximado de I mediante la fÃ³rmula de Simpson = ', simpson(a,b,fa,fpm,fb) ) 
```

```python
# MÃ©todos simples de integraciÃ³n numÃ©rica (hecha por madra)

def puntoMedio(a, b, f):
    aprox_pm = (b-a) * f((a + b)/2)
    return aprox_pm

def trapecio(a, b, f):
    aprox_tr = (b-a) * (f(a) + f(b))/2
    return aprox_tr

def simpson(a, b, f):
    aprox_simp = (b-a) * (f(a) + 4*f((a + b) / 2) + f(b))/6
    return aprox_simp

x = sp.Symbol('x', real = True)

f_exp = x**4 + 1
f = sp.lambdify(x,f_exp)

a = 0
b = 3

print('Valor aproximado de I mediante la fÃ³rmula del punto medio = ', puntoMedio(a,b,f) ) 
print('Valor aproximado de I mediante la fÃ³rmula del trapecio = ', trapecio(a,b,f) ) 
print('Valor aproximado de I mediante la fÃ³rmula de Simpson = ', simpson(a,b,f) ) 
```

```python
# MÃ©todos compuestos de integraciÃ³n numÃ©rica (hecha por udc)

x = sp.Symbol('x', real = True)
f_exp = x**4+1
f = sp.lambdify(x,f_exp)

a = 0; b = 3
n = 100

x1 = np.linspace(a,b,n+1) # aquÃ­ guardamos los x_{i}. 
                          # Recuerda que, en Python, se guarda x1[0], x1[1], ..., x1[(n+1)-1] = x1[n]
y1 = f(x1)

h = (b-a)/n # el tamaÃ±o de cada subintervalo

aprox_medio = 2*h * np.sum(y1[1:n:2])
aprox_trap = h/2 * (y1[0]+2*np.sum(y1[1:n])+y1[n])
aprox_simpson = 2*h/6 * (y1[0] + 4*np.sum(y1[1:n:2])+2*np.sum(y1[2:n-1:2])+y1[n])

print('aprox_medio: ',aprox_medio) 
print('aprox_trap: ',aprox_trap) 
print('aprox_simpson: ',aprox_simpson) 

print('Exacta: ',b**5/5+b)
```

```python
# MÃ©todos compuestos de integraciÃ³n numÃ©rica (hecha por madra)

def puntoMedioCompuesto(f, a, b, n):
    if n <= 0 or not isinstance(n, int):
        raise ValueError("n debe ser un entero positivo")

    h = (b - a) / n
    xMedios = np.linspace(a, b, n+1)
    yMedios = f(xMedios)

    aprox = 2*h * np.sum(yMedios[1:n:2])
    return aprox

def trapecioCompuesto(f, a, b, n):
    if n <= 0 or not isinstance(n, int):
        raise ValueError("n debe ser un entero positivo")

    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)

    aprox = h / 2 * (y[0] + 2 * np.sum(y[1:n]) + y[n])
    return aprox

def simpsonCompuesto(f, a, b, n):
    if n <= 0 or not isinstance(n, int):
        raise ValueError("n debe ser un entero positivo")

    if n % 2 != 0:
        raise ValueError("Para Simpson compuesto, n debe ser PAR")

    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    y = f(x)

    aprox = h / 3 * (
        y[0]
        + 4 * np.sum(y[1:n:2])
        + 2 * np.sum(y[2:n-1:2])
        + y[n]
    )
    return aprox


x = sp.Symbol('x', real=True)
f_exp = x**4 + 1
f = sp.lambdify(x, f_exp, "numpy")

a = 0
b = 3
n = 100

print("Punto medio:  ", puntoMedioCompuesto(f, a, b, n))
print("Trapecio:     ", trapecioCompuesto(f, a, b, n))
print("Simpson:      ", simpsonCompuesto(f, a, b, n))

print("Exacta:       ", b**5/5 + b)
```

```python
# Integrales impropias

# --- Ejemplo 1 ---

import sympy as sp
x = sp.symbols('x', real = True)
M = sp.Symbol('M', real = True)

f_exp = sp.exp(x)

# CÃ¡lculo directo
I_directo = sp.integrate(f_exp,(x,-sp.oo,0))
print('Integral de ',f_exp,' entre -oo y 0 es = ',I_directo)

# CÃ¡lculo con lÃ­mites
I_limites = sp.limit( sp.integrate(f_exp,(x,-M,0)), M, +sp.oo )
print('Integral de ',f_exp,' entre -oo y 0 es = ',I_limites)

# ---- Ejemplo 2 ----

import sympy as sp
x = sp.symbols('x', real = True)
c = sp.Symbol('c', real = True)

f_exp = 1/sp.sqrt(x)

# CÃ¡lculo directo
I_directo = sp.integrate(f_exp,(x,0,2))
print('La integral vale = ', I_directo)

# CÃ¡lculo con lÃ­mites
I_limites = sp.limit( sp.integrate(f_exp,(x,c,2)), c, 0, dir='+')
print('La integral vale = ', I_limites)

# ---- Ejemplo 3 ----

import sympy as sp
x = sp.symbols('x', real = True)
c = sp.Symbol('c', real = True)

f_exp = 1/x

# CÃ¡lculo con lÃ­mites
I_limites = sp.limit( sp.integrate(f_exp,(x,c,2)), c, 0, dir='+')
print('La integral vale = ', I_limites)
```
