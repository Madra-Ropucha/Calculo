```python
#Zieba Glaz Szymon Artur

import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
# EJERCICIO 2

x = sp.symbols('x')
gExpr = sp.sqrt(5 - x)

g = sp.Lambda(x, gExpr)

a_taylor = 1
n_grado = 4

pol_taylor = sp.series(gExpr, x, a_taylor, n_grado + 1).removeO()
print(f"\nPolinomio de Taylor (grado {n_grado}, en a={a_taylor}):")
print(f"  P(x) = {pol_taylor}")
pol_taylor

valor_taylor = pol_taylor.subs(x, sp.sqrt(2))
print(f"\nValor del polinomio de Taylor en x=sqrt(2): {valor_taylor}")


a_taylor = 1
n_grado = 4

pol_taylor = sp.series(gExpr, x, a_taylor, n_grado + 1).removeO()

valor_taylor = pol_taylor.subs(x, sp.sqrt(2))

sp.simplify(pol_taylor.subs(x, sp.sqrt(2)))

# punto donde evaluamos el error
x_val = sp.sqrt(2)

# derivada de orden 5 de g
g5 = sp.diff(gExpr, x, 5)

# cota M: máximo del valor absoluto de g5 en el intervalo [a_taylor, x_val]
# evaluamos en los extremos (para esta función la magnitud es monótona en el intervalo)
abs_at_a = abs(sp.N(g5.subs(x, a_taylor)))
abs_at_x = abs(sp.N(g5.subs(x, x_val)))
M = max(abs_at_a, abs_at_x)

# cota del resto de Lagrange: M / 5! * |x - a|^5
resto_cota = M / sp.factorial(5) * abs(x_val - a_taylor)**5

print("g^(5)(x) =", sp.simplify(g5))
print(f"|g^(5)| en a={a_taylor}       :", abs_at_a)
print(f"|g^(5)| en x={x_val} :", abs_at_x)
print("Cota M usada :", M)
print("Cota del resto (Lagrange) en x=sqrt(2) =", float(sp.N(resto_cota)))

sp.simplify(pol_taylor)
sp.simplify(valor_taylor)
```

```python
#EJERCICIO 3

x = sp.symbols('x')
t = sp.symbols('t')

fDenExpr = 5 + sp.atan(x)

# apartado A
# Calcular los valores donde f(x) = 0
soluciones = sp.solve(fDenExpr, x)
print("Soluciones de fDenExpr = 0:", soluciones)

# x, 1, |x| estan definidas en todo R

fExpr = 1 / fDenExpr + x + sp.Abs(x)
f = sp.Lambda(x, fExpr)
f(x)


print("\nLimites en x -> 0:")
print(sp.limit(f(x), x, 0, dir='-'))
print(sp.limit(f(x), x, 0))
print(sp.limit(f(x), x, 0, dir='+'))

#limites

print(sp.limit(f(x), x, -sp.oo))
print(sp.limit(f(x), x, sp.oo))

m = sp.limit(f(t)/t, t, sp.oo)
n = sp.limit(f(t) - m*t, t, sp.oo)

print(f"resultado de m: {m}")
print(f"resultado de n: {n}")

print(sp.limit(f(t) - (m*t +n), t, -sp.oo))
print(sp.limit(f(t) - (m*t +n), t, sp.oo))

t_vals = np.linspace(-0.25, 0.25, 5000)
f_vals = [f(t_val) for t_val in t_vals]

plt.plot(t_vals, f_vals, label="f(t)")
plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
plt.title("Plot of f(t)")
plt.xlabel("t")
plt.ylabel("f(t)")
plt.legend()
plt.grid()
plt.show()

f(x)
```
