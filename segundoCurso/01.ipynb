```python
import sympy as sp
import numpy as np
```

```python
# Deficion de una funcion a trozos

x = sp.symbols('x')
fExpr = sp.Piecewise((1/(x), x>0), (1, x<0))
display(fExpr)

#TODO: mirar que es lo que hace la funcion HeavySide
#fExpr = sp.Heaviside(x)
#display(fExpr)
```

```python
f = sp.Lambda(x, x/(sp.cos(x)))
display(sp.calculus.singularities(f(x), x))
```

```python
# Calcula los puntos en los que hay division por cero

f = sp.Lambda(x, sp.sqrt(x**2 - 1))
display(f(x))
display(sp.calculus.singularities(f(x), x))
```

```python
# Calcula el rango de la imagen de la funcion

f = sp.Lambda(x, sp.sqrt(x**2 - 1))
R = sp.calculus.util.function_range(f(x), x, sp.S.Reals)
display(R)
```

```python
# Calculo de limites

gExpr = sp.cos(x)/(x + 1)
g = sp.Lambda(x, gExpr)
display(g(x))

# En el caso de que no se especifique la direccion, sympy calcula el limite por la derecha
display(sp.limit(g(x), x, -1))

display(sp.limit(g(x), x, -1, dir='-'))
display(sp.limit(g(x), x, -1, dir='+'))
```

```python
# Calculo de limites en el infinito

gExpr = sp.cos(x)/(x + 1)
g = sp.Lambda(x, gExpr)
display(g(x))

display(sp.limit(g(x), x, sp.oo))
display(sp.limit(g(x), x, -sp.oo))
```

```python
# Calculo del dominio de continuidad

f = sp.Lambda(x, sp.abs(x))
I = sp.calculus.util.continuous_domain(f, x, sp.Reals)
display(I)
```
