```python
import sympy as sp
```

```python
x = sp.symbols('x')
fExpr = sp.ln(x + 2) + sp.ln(x - 2) 
gExpr = 1

equation = sp.Eq(fExpr, gExpr)
display(equation)

display(sp.solve(equation, x))
display(sp.N(sp.solve(equation, x)[0])) # Solucion numerica
```

```python
x = sp.symbols('x')
fExpr = sp.Abs(5*x - 2) - 2
display(fExpr)
```

```python
x = sp.symbols('x')
numerador = 3*x - sp.tan(2*x)
denominador = sp.sqrt(1 - x**2)
fExpr = numerador/denominador
f = sp.Lambda(x, fExpr)
display(f(x))

display(sp.solve((denominador)**2>=0, x))
display(sp.solve(denominador, x))
display(sp.calculus.util.continuous_domain(fExpr, x, sp.Reals))
```

```python
x = sp.symbols('x')
f2 = sp.Lambda(x, x*2**3 - x)
print('f2 es impar?')
print(f2(x) == -f2(-x))
print('f2 es par?')
print(f2(x) == f2(-x))
```
