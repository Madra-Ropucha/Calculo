```python
import sympy as sp

x = sp.symbols('x', real=True)
expr = sp.Abs(2*x-4)-6
display(expr)
soluciones = sp.solve(expr)
display(soluciones)
print("Solucion: ")

for solucion in soluciones:
    print(solucion)

p = sp.plot(sp.Abs(2*x-4)-6, (x, -2, 6), show=False)
p[0].line_color='r'
p.xlabel='x'
p.ylabel='y'
p.legend=True
p.show()
```

```python
# Ejercicio 15 apartado 4

import sympy as sp

x = sp.symbols('x', real=True)
f_expr = (sp.tan(x) + sp.sin(x)) / sp.cos(x**2)
display(f_expr)
f = sp.Lambda((x), f_expr)
print("Es par? ", f(x) == f(-x))
print("Es impar? ", -f(x) == f(-x).simplify())
```

```python
# Ejercicio 15 apartado 4

import sympy as sp

x = sp.symbols('x', real=True)
f_expr = (x**4 * sp.sin(x)**3) / sp.tan(x)
display(f_expr)
f = sp.Lambda((x), f_expr)
print("Es par? ", f(x).simplify() == f(-x).simplify())
print("Es impar? ", -f(x).simplify() == f(-x).simplify())

display(f(x))
display(f(-x))
display(-f(x).simplify())
```

```python
%matplotlib inline

import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x', real=True)
expr = (sp.tan(x) + sp.sin(x)) / sp.cos(x**2)
display(expr)
soluciones = sp.solve(expr)
display(soluciones)
print("Solucion: ")

for solucion in soluciones:
    print(solucion)

p = sp.plot(expr, (x, -5, 6), show=False)
p[0].line_color='r'
p.xlabel='x'
p.ylabel='y'
p.legend=True
p.show()
```

```python
import sympy as sp
import matplotlib.pyplot as plt

%matplotlib inline

x = sp.symbols('x', real=True)
f_expr = 6*x/(x**2-9)
display(f_expr)
f = sp.Lambda((x),f_expr)
g_expr = sp.sqrt(3*x) # (3*x)**(1/2)
display(g_expr)
g = sp.Lambda((x),g_expr)
display(f(g(12)))
display(f(g(x)))
display(sp.calculus.util.continuous_domain(f(g(x)), x, sp.Reals))
```

```python
%matplotlib inline

import sympy as sp

x = sp.symbols('x', real=True)
u = sp.Symbol('u', real = True)
f_expr = (u-1)/(u+1)
display(f_expr)
f = sp.Lambda((u),f_expr)
g_expr = (u+1)/(1-u)
display(g_expr)
g = sp.Lambda((u),g_expr)
 
display(f(g(u)).simplify())
display(g(f(u)).simplify())
 
p = sp.plot(f_expr, g_expr, u, (u, -0.9, 0.9), show=False)
p[0].line_color='r'
p[1].line_color='b'
p.xlabel='x'
p.ylabel='y'
p.legend=True
p.show()
```
