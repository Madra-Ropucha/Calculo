```python
import sympy as sp

x = sp.Symbol('x', real=True)

fExpr = sp.E**x * sp.ln(x) + x**3 - 2
fExprDiff1 = sp.diff(fExpr, x, 1)
fExprDiff2 = sp.diff(fExpr, x, 2)
fExprDiff3 = sp.diff(fExpr, x, 3)

display(fExpr)
display(fExprDiff1)
display(fExprDiff2)
display(fExprDiff3)

print(fExpr.subs(x, 1.5))
print(fExprDiff1.subs(x, 1.5))

#display(fExpr.solve)

display(sp.solve(fExpr, x))

display(x - fExpr / fExprDiff1)
print(1,5 - fExpr.subs(x, 1.5) / fExprDiff1.subs(x, 1.5))

p = sp.plot(sp.E**x * sp.ln(x) + x**3 - 2, (x, -4, 4), show=False)
p[0].line_color='r'
p.xlabel='x'
p.ylabel='y'
p.show()
```
