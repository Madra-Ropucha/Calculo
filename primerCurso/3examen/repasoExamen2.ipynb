```python
# Metodo de dicotomia
#
# Sirve para aproximar la raiz de una funcion

import numpy as np
import sympy as sp
x = sp.symbols('x', real=True) 

def Dicotomia(a, b, fExpr, tol = 1.e-5):
    f = sp.Lambda(x, fExpr)

    N_max = 100

    x_aprox = np.zeros(N_max)

    for k in range(0,N_max):
        x_aprox[k] = (a+b) / 2

        if f(x_aprox[k]) == 0: break

        if f(a) * f(x_aprox[k]) < 0:
            b = x_aprox[k]
        else:
           a = x_aprox[k]

        err_relativo = np.abs( x_aprox[k]-x_aprox[k-1] ) / np.abs( x_aprox[k] )
        if ( (k > 0) and ( err_relativo < tol ) ): break
        
    return x_aprox[:np.argmax(x_aprox == 0)] if 0 in x_aprox else x_aprox

print(Dicotomia(-2., 2., x**3 + 2))


# import numpy as np
# import sympy as sp

# x = sp.symbols('x', real=True) 
# f_expr = x**3 + 2
# f = sp.Lambda(x,f_expr)

# N_max = 100
# tol = 1.e-5
# a = -2.
# b = 2.

# x_aprox = np.zeros(N_max)

# for k in range(0,N_max):
#     x_aprox[k] = (a+b) / 2

#     if f(x_aprox[k]) == 0: break
      
#     if f(a) * f(x_aprox[k]) < 0:
#         b = x_aprox[k]
#     else:
#        a = x_aprox[k]

#     err_relativo = np.abs( x_aprox[k]-x_aprox[k-1] ) / np.abs( x_aprox[k] )
#     if ( (k > 0) and ( err_relativo < tol ) ): break

# print('Número de iteraciones realizadas: ', k+1) 
#       # NOTA: Contamos 1 más, k+1, porque empezamos el bucle en 0
# print('Aproximación de la raíz: ', x_aprox)
    
```

```python
def f(x):
    return x**3 + 2

# Initial interval
a, b = -2, 2

# Perform 10 iterations of the bisection method
for i in range(100):
    # Calculate midpoint
    c = (a + b) / 2
    
    # Update the interval based on the sign of f(c)
    if f(c) * f(a) < 0:
        b = c  # Root is in the left subinterval
    else:
        a = c  # Root is in the right subinterval

# Print the final midpoint after 10 iterations, which approximates the root
root_approx = (a + b) / 2
print(f"Approximate root after 10 iterations: {root_approx}")
```
