### Ejercicio 1

```python
import sympy as sp

x = sp.Symbol('x', real = True)
y = sp.Symbol('y', real = True)

fExpr = sp.Abs(x**2 + 1)
gExpr = sp.Abs(3 - 2*x)
hExpr = (x - 4)**2
iExpr = sp.Abs(2 - x**2)

display(fExpr)
display(gExpr)
display(hExpr)
display(iExpr)

f = sp.Lambda(x, fExpr)
g = sp.Lambda(x, gExpr)
h = sp.Lambda(x, hExpr)
i = sp.Lambda(x, iExpr)

p = sp.plot(
            # f(x), 2,
            # g(x), 3,
            # h(x), 4,
            i(x), 1,
            show = False
        )
p.line_color = 'r'
p.xlabel = 'x'
p.ylabel = 'y'
p.xlim = (-2, 2)
p.ylim = (-3, 3)
p.legend = True
p.show()
```

### Ejercicio 2

```python
import sympy as sp

x = sp.Symbol('x', real = True)
y = sp.Symbol('y', real = True)

fExpr = sp.sin(x**2)
gExpr = sp.Abs(x)*(x + 1)**2
hExpr = (x * sp.Abs(x)) / (1 + sp.Abs(x))

display(fExpr)
display(gExpr)
display(hExpr)

f = sp.Lambda(x, fExpr)
g = sp.Lambda(x, gExpr)
h = sp.Lambda(x, hExpr)

p = sp.plot(f(x), g(x), h(x), show = False)
p.line_color = 'r'
p.xlabel = 'x'
p.ylabel = 'y'
p.xlim = (-3, 3)
p.ylim = (-3, 3)
p.legend = True
p.show()
```

### Ejercicio 3

```python
import sympy as sp

x = sp.Symbol('x', real = True)
y = sp.Symbol('y', real = True)

fExpr = sp.ln((x + 1) / (x - 1))
gExpr = 1 / (sp.tanh(x / 2))

display(fExpr)
display(gExpr)

f = sp.Lambda(x, fExpr)
g = sp.Lambda(x, gExpr)

p = sp.plot(f(x), g(x), show = False)
p.line_color = 'r'
p.xlabel = 'x'
p.ylabel = 'y'
p.xlim = (-8, 8)
p.ylim = (-8, 8)
p.legend = True
p.show()
```

### Ejercicio 4

```python
import sympy as sp

x = sp.Symbol('x', real = True)
y = sp.Symbol('y', real = True)

fExpr = 3 + sp.E**(x - 4)
gExpr = sp.ln(x - 3) + 4

display(fExpr)
display(gExpr)

f = sp.Lambda(x, fExpr)
g = sp.Lambda(x, gExpr)

display(f(g(x)))
display(g(f(x)))

p = sp.plot(f(x), g(x), show = False)
p.line_color = 'r'
p.xlabel = 'x'
p.ylabel = 'y'
p.xlim = (-5, 10)
p.ylim = (-2, 10)
p.legend = True
p.show()

print(sp.ln(0).evalf())
```
