```python
#Formulas compuestas: Pto medio, Simpson y trapecio

import sympy as sp
import numpy as np

x = sp.Symbol('x', real = True)
f_exp = sp.sin(x) / x
I = sp.integrate(f_exp,(x, sp.pi, sp.pi * 2))
display(I)
f = sp.lambdify(x,f_exp)

a = 3.14
b = 2 * 3.14
n = 100 #el numero de subintervalos menos 1. Para Simpson compuesto y para el punto medio compuesto, este numero tiene que ser par

x1 = np.linspace(a,b,n+1) # aqu√≠ guardamos los x_{i}. 
#print(x1)                 # Recuerda que, en Python, se guarda x1[0], x1[1], ..., x1[(n+1)-1] = x1[n]
y1 = f(x1)
#print(y1)
h = (b-a)/n # el tama√±o de cada subintervalo

aprox_simpson = h/3 * (y1[0] + 4*np.sum(y1[1:n:2])+2*np.sum(y1[2:n-1:2])+y1[n])

print('Valor aproximado de I mediante la f√≥rmula de Simpson compuesto: ',aprox_simpson)
```
