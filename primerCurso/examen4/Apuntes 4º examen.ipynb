```python
#4.4 Integracion numerica

#Formulas simples: Pto medio, Simpson y trapecio

import sympy as sp
import numpy as np

def pto_medio(a, b, fpm):
    aprox_pm = (b-a) * fpm
    return aprox_pm

def trapecio(a, b, fa, fb):
    aprox_tr = (b-a) * (fa + fb)/2
    return aprox_tr

def simpson(a, b, fa, fpm, fb):
    aprox_simp = (b-a) * (fa + 4*fpm + fb)/6
    return aprox_simp

x = sp.Symbol('x', real = True)

f_exp = x**4 + 1
f = sp.lambdify(x,f_exp)

a = 2
b = 6
pm = (a+b)/2

fa = f(a)
fpm = f(pm)
fb = f(b)

print('Valor aproximado de I mediante la fórmula del punto medio = ', pto_medio(a,b,fpm) ) 
print('Valor aproximado de I mediante la fórmula del trapecio = ', trapecio(a,b,fa,fb) ) 
print('Valor aproximado de I mediante la fórmula de Simpson = ', simpson(a,b,fa,fpm,fb) ) 
```

```python
#Formulas compuestas: Pto medio, Simpson y trapecio

import sympy as sp
import numpy as np

x = sp.Symbol('x', real = True)
f_exp = sp.sin(x)
I = sp.integrate(f_exp,(x, 0, sp.pi))
display(I)
f = sp.lambdify(x,f_exp)

a = sp.pi
b = 2 * sp.pi
n = 5 #el numero de subintervalos menos 1. Para Simpson compuesto y para el punto medio compuesto, este numero tiene que ser par

x1 = np.linspace(a,b,n+1) # aquí guardamos los x_{i}. 
print(x1)                 # Recuerda que, en Python, se guarda x1[0], x1[1], ..., x1[(n+1)-1] = x1[n]
y1 = f(x1)
print(y1)
h = (b-a)/n # el tamaño de cada subintervalo

aprox_medio = 2*h * np.sum(y1[1:n:2])
aprox_trap = h/2 * (y1[0]+2*np.sum(y1[1:n])+y1[n])
aprox_simpson = h/3 * (y1[0] + 4*np.sum(y1[1:n:2])+2*np.sum(y1[2:n-1:2])+y1[n])

print('Valor aproximado de I mediante la fórmula del punto medio compuesto: ',aprox_medio)
print('Valor aproximado de I mediante la fórmula del trapecio compuesto: ',aprox_trap)  
print('Valor aproximado de I mediante la fórmula de Simpson compuesto: ',aprox_simpson)
```

```python
#4.6 Aplicaciones de la Integral
#Cálculo de áreas

import sympy as sp
x = sp.Symbol('x', real=True)
r = sp.Symbol('r', positive=True)
t = sp.Symbol('t')
f_exp = sp.sqrt(r**2-x**2)
A = 2*sp.simplify(sp.integrate(f_exp,(x,-r,r)))
print('El area del circulo de radio r es: ')
display(A)
```

```python
# Calcular el área de la superficie comprendida entre las gráficas de y=x^3 e y=x

import sympy as sp
x=sp.symbols('x')
f_exp=x**3
g_exp=x
sp.plot((f_exp,(x,-1,1)),(g_exp,(x,-1,1)))
roots=sp.solve(f_exp-g_exp)
print('Puntos de corte: ',roots)
A=sp.integrate(f_exp-g_exp,(x,-1,0))+sp.integrate(g_exp-f_exp,(x,0,1))
print('El area de la sup comprendida entre las dos graficas es: ',A)
```

```python
#Cálculo de volúmenes de sólidos de revolución

# Calcular el volumen de una esfera de radio r.

import sympy as sp
x=sp.symbols('x')
r = sp.Symbol('r', positive=True)
f=sp.sqrt(r**2-x**2)
V=sp.pi*sp.integrate(f**2,(x,-r,r))
print('El volumen del sólido de revolución es: ')
display(V)
```

```python
# Calcular el volumen del sólido de revolución generado al rotar la superficie comprendida entre las gráficas de 
# y=x^3 e y=x alrededor del eje X.

import sympy as sp
x=sp.symbols('x')
f=x**3
g=x
roots=sp.solve(f-g)
print('Puntos de corte: ',roots)
V=sp.pi*sp.integrate(g**2-f**2,(x,-1,0))+sp.pi*sp.integrate(g**2-f**2,(x,0,1))
print('El volumen es: ',V)
```

```python
# Calcular el volumen del sólido generado al rotar alrededor del eje 0Y la región comprendida entre la parábola 
# entre y=x^2 e y=0 e y=4 

import sympy as sp
x,y=sp.symbols('x,y')
sp.plot((x**2,(x,0,2)),(4,(x,0,2)))
f=sp.sqrt(y)
V=sp.pi*sp.integrate(f**2,(y,0,4))
print('El volumen del solido de revolucion es: ',V)
```

```python
# Limites

import sympy as sp

x = sp.Symbol('x')

limit = sp.limit(x**2 - 1, x, 1)
print(limit)  # Salida: 0

limit = sp.limit(1/x, x, sp.oo)
print(limit)  # Salida: 0

limit_izq = sp.limit(1/x, x, 0, '+')  # Límite por la derecha
limit_der = sp.limit(1/x, x, 0, '-')  # Límite por la izquierda
print(limit_izq, limit_der)  # Salida: oo, -oo
```

```python
# Derivacion

import sympy as sp

# Definimos la variable simbólica x
x = sp.Symbol('x')

# Definimos la función f(x)
f = x**2 + 2*x + 1

# Calculamos la derivada de f respecto a x
dfdx = sp.diff(f, x)

print(dfdx)  # Imprime el resultado: 2*x + 2
```
